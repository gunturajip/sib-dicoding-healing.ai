```python
tmp_taman = pd.DataFrame()
tmp_museum = pd.DataFrame()
```

# Taman Surabaya

```python
tmp = taman_surabaya.copy()
tmp.drop(columns = ["id", "Keyword", "Website", "Plus_Code", "Reviews"], inplace=True)
tmp.dropna(subset = ["Name", "Address", "Rating"], inplace=True)
tmp["Provinsi"] = tmp["Address"].apply(lambda data: data.split(",")[::-1][0].strip())
tmp["Kota"] = tmp["Address"].apply(lambda data: data.split(",")[::-1][1].strip())
tmp["Kecamatan"] = tmp["Address"].apply(lambda data: data.split(",")[::-1][2].strip())
tmp.head()
```

```python
tmp["Provinsi"] = tmp["Provinsi"].apply(lambda data: " ".join(data.split()[:2]))
tmp["Provinsi"].replace("East Java", "Jawa Timur", inplace=True)
tmp["Provinsi"].astype("category")
```

```python
tmp.drop(tmp.loc[tmp["Kota"] == "Kabupaten Sidoarjo"].index, inplace=True)
tmp["Kota"].replace(["Kota SBY", "Surabaya City"], "Surabaya", inplace=True)
tmp["Kota"].astype("category")
```

```python
tmp["Kecamatan"] = tmp["Kecamatan"].apply(lambda data: " ".join(data.split()[1:]) if (len(data.split())>1 and "Kec" in data.split()[0]) else data)
tmp["Kecamatan"].astype("category")
```

```python
try:
    tmp["Rating"] = tmp["Rating"].str.replace(",", ".").astype(float)
except:
    print("Udah Benar")
tmp.head()
```

```python
print(tmp.isnull().sum())
tmp_new = pd.DataFrame()
tmp_new["nama"] = tmp["Name"]
tmp_new["kategori"] = "Taman"
tmp_new["rating"] = tmp["Rating"]
tmp_new["provinsi"] = tmp["Provinsi"]
tmp_new["kota"] = tmp["Kota"]
tmp_new["kecamatan"] = tmp["Kecamatan"]
tmp_new["alamat"] = tmp["Address"]
tmp_new["link"] = tmp["URL"]
tmp_taman = tmp_taman.append(tmp_new, ignore_index=True)
```

```python
# Mengimpor library TfidfVectorizer untuk pemodelan Content-Based Filtering
from sklearn.feature_extraction.text import TfidfVectorizer
 
# Inisialisasi TfidfVectorizer
tf = TfidfVectorizer()
 
# Melakukan perhitungan idf pada kolom kecamatan
tf.fit(taman_cbf['kecamatan'])
 
# Mendata penerbit buku pada data taman_cbf
tf.get_feature_names()

# Melakukan fit lalu ditransformasikan ke bentuk dataframe
tfidf_matrix = tf.fit_transform(taman_cbf['kecamatan'])
 
# Melihat ukuran dataframe tfidf
tfidf_matrix.shape

# Mengubah vektor tf-idf dalam bentuk matriks dengan fungsi todense()
tfidf_matrix.todense()

# Membuat dataframe untuk melihat tf-idf matrix
# Baris diisi dengan judul buku
# Kolom diisi dengan penerbit buku

pd.DataFrame(
    tfidf_matrix.todense(),
    columns=tf.get_feature_names(),
    index=taman_cbf['nama']
).sample(5, axis=0).sample(5, axis=1)
```

```python
# Mengimpor library cosine_similarity untuk pemodelan Content-Based Filtering
from sklearn.metrics.pairwise import cosine_similarity
 
# Menghitung cosine similarity pada matrix tf-idf
cosine_sim = cosine_similarity(tfidf_matrix)

# Menampilkan array hasil komputasi fungsi cosine_similarity
cosine_sim
```

```python
# Membuat dataframe dari variabel cosine_sim dengan baris dan kolom berupa nama
cosine_sim_df = pd.DataFrame(cosine_sim, index=taman_cbf['nama'], columns=taman_cbf['nama'])

# Menampilkan bentuk data cosine_sim_df
print('Bentuk :', cosine_sim_df.shape)
 
# Melihat similarity matrix pada setiap buku
cosine_sim_df.sample(5, axis=0).sample(5, axis=1)
```

```python
# Membuat fungsi rekomendasi data untuk Content-Based Filtering
def rekomendasi_taman_cbf(nama, similarity_data=cosine_sim_df, items=taman_cbf[['nama', 'kategori', 'rating', 'kecamatan', 'kota', 'provinsi', 'alamat', 'link']], k=20):
    """
    Rekomendasi data berdasarkan kemiripan dataframe
 
    Parameter:
    ---
    nama : tipe data string (str)
                Nama data (index kemiripan dataframe)
    similarity_data : tipe data pd.DataFrame (object)
                      Kesamaan dataframe, simetrik, dengan nama data sebagai 
                      indeks dan kolom
    items : tipe data pd.DataFrame (object)
            Mengandung kedua nama dan fitur lainnya yang digunakan untuk mendefinisikan kemiripan
    k : tipe data integer (int)
        Banyaknya jumlah rekomendasi yang diberikan
    ---
 
 
    Pada index ini, kita mengambil k dengan nilai similarity terbesar 
    pada index matrix yang diberikan (i).
    """

    # Mengambil data dengan menggunakan argpartition untuk melakukan partisi secara tidak langsung sepanjang sumbu yang diberikan    
    # Dataframe diubah menjadi numpy
    # Range(start, stop, step)
    index = similarity_data.loc[:,nama].to_numpy().argpartition(range(-1, -k, -1))
    
    # Mengambil data dengan similarity terbesar dari index yang ada
    closest = similarity_data.columns[index[-1:-(k+2):-1]]
    
    # Drop nama agar nama data yang dicari tidak muncul dalam daftar rekomendasi
    closest = closest.drop(nama, errors='ignore')
 
    return pd.DataFrame(closest).merge(items).head(k)
```

```python
# Mengimpor modul train_test_split dari library sklearn untuk pendukung pembagian data
from sklearn.model_selection import train_test_split

# Mengacak data untuk Collaborative Filtering dari data asli
df_cf = df.sample(frac=1, random_state=123)

# Menggunakan seperseratus data asli untuk Collaborative Filtering
# Menampung pada variabel df_cf
df_cf = df_cf[:len(df)//10]

# Menampilkan jumlah data untuk pemodelan berbasis Collaborative Filtering
print(len(df_cf))

# Mendata nilai fitur id_pengguna yang unik
id_pengguna = df_cf['id_pengguna'].unique().tolist()

# Mendata nilai fitur id_buku yang unik
id_buku = df_cf['id_buku'].unique().tolist()

# Memetakan id_pengguna dan indeks mulai 0 hingga jumlah total data id_pengguna
id_pengguna_dan_indeks = {id_pengguna: indeks for indeks, id_pengguna in enumerate(id_pengguna)}

# Memetakan id_buku dan indeks mulai 0 hingga jumlah total data id_buku
id_buku_dan_indeks = {id_buku: indeks for indeks, id_buku in enumerate(id_buku)}

# Memetakan indeks mulai 0 hingga jumlah total data id_pengguna dan id_pengguna
indeks_dan_id_pengguna = {indeks: id_pengguna for indeks, id_pengguna in enumerate(id_pengguna)}

# Memetakan indeks mulai 0 hingga jumlah total data id_buku dan id_buku
indeks_dan_id_buku = {indeks: id_buku for indeks, id_buku in enumerate(id_buku)}

# Menambahkan fitur indeks_id_pengguna yang merupakan hasil pemetaan id_pengguna menjadi indeks yang unik
df_cf['indeks_id_pengguna'] = df_cf['id_pengguna'].map(id_pengguna_dan_indeks)

# Menambahkan fitur indeks_id_buku yang merupakan hasil pemetaan id_buku menjadi indeks yang unik
df_cf['indeks_id_buku'] = df_cf['id_buku'].map(id_buku_dan_indeks)

# Menampung jumlah id_pengguna unik
jumlah_pengguna = len(id_pengguna_dan_indeks)

# Menampilkan jumlah pengguna
print(jumlah_pengguna)

# Menampung jumlah id_buku unik
jumlah_buku = len(id_buku_dan_indeks)

# Menampilkan jumlah buku
print(jumlah_buku)

# Menampung nilai minimal dari rating_buku pada data df_cf
min_rating = min(df_cf['rating_buku'])

# Menampung nilai maksimal dari rating_buku pada data df_cf
max_rating = max(df_cf['rating_buku'])

# Membuat variabel x untuk mencocokkan data pengguna dan buku menjadi satu value
X = df_cf[['indeks_id_pengguna', 'indeks_id_buku']].values
 
# Membuat variabel y untuk membuat kemiripan rating pada rentang 0-1
y = df_cf['rating_buku'].apply(lambda x: (x - min_rating) / (max_rating - min_rating)).values
 
# Membagi masukan (X) dan hasil masukan (y) dengan persentase data validasi 20%
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2)

# Menampilkan bentuk fitur masukan pada data latih Collaborative Filtering
print(X_train.shape)

# Menampilkan bentuk fitur target pada data latih Collaborative Filtering
print(y_train.shape)
```

```python
# Import library tensorflow untuk pengembangan Deep Learning
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

# Membuat kelas RecommenderNet untuk pemodelan terkait sistem rekomendasi
class RecommenderNet(tf.keras.Model):
 
  # Inisialisasi parameter kelas
  def __init__(self, jumlah_pengguna, jumlah_buku, ukuran_embedding, **kwargs):

    # Pewarisan parameter - paramter pada kelas
    super(RecommenderNet, self).__init__(**kwargs)

    # Inisialisasi variabel kelas
    self.jumlah_pengguna = jumlah_pengguna
    self.jumlah_buku = jumlah_buku
    self.ukuran_embedding = ukuran_embedding

    # Inisialisasi layer embedding terhadap data pengguna
    self.embedding_pengguna = layers.Embedding(
        jumlah_pengguna,
        ukuran_embedding,
        embeddings_initializer='he_normal',
        embeddings_regularizer=keras.regularizers.l2(1e-6)
    )

    # Inisialisasi layer embedding terhadap bias pada data pengguna
    self.bias_pengguna = layers.Embedding(jumlah_pengguna, 1)

    # Inisialisasi layer embedding terhadap data buku
    self.embedding_buku = layers.Embedding(
        jumlah_buku,
        ukuran_embedding,
        embeddings_initializer='he_normal',
        embeddings_regularizer=keras.regularizers.l2(1e-6)
    )

    # Inisialisasi layer embedding terhadap bias pada data buku
    self.bias_buku = layers.Embedding(jumlah_buku, 1)
 
  # Inisialisasi fungsi pemanggilan input bagi model Deep Learning
  def call(self, inputs):

    # Inisialisasi layer embedding untuk vektor pengguna
    vektor_pengguna = self.embedding_pengguna(inputs[:,0])

    # Inisialisasi layer embedding untuk bias_pengguna
    bias_pengguna = self.bias_pengguna(inputs[:,0])

    # Inisialisasi layer embedding untuk vektor_buku
    vektor_buku = self.embedding_buku(inputs[:,1])

    # Inisialisasi layer embedding untuk bias_buku
    bias_buku = self.bias_buku(inputs[:,1])

    # Menghitung perkalian matriks vektor_pengguna dan vektor_buku pada axes 2
    titik_pengguna_dan_buku = tf.tensordot(vektor_pengguna, vektor_buku, 2) 

    # Menjumlahkan hasil perkalian matriks dengan bias_pengguna dan bias_buku pada variabel x
    x = titik_pengguna_dan_buku + bias_pengguna + bias_buku
    
    # Melakukan aktivasi dengan fungsi sigmoid pada variabel x
    return tf.nn.sigmoid(x)
```

```python


# Membuat model berbasis RecommenderNet dengan parameter yang telah dibuat
model = RecommenderNet(jumlah_pengguna, jumlah_buku, 50)
 
# Mengompile model dengan optimizer, loss, dan metrik yang relevan
model.compile(
    loss='binary_crossentropy',
    optimizer='adam',
    metrics=[tf.keras.metrics.RootMeanSquaredError()]
)
```

```python
# Melatih model dengan epoch 10 dan batch_size 8 dan data latih serta data validasi
history = model.fit(
    x=X_train,
    y=y_train,
    batch_size=8,
    epochs=10,
    validation_data=(X_val,y_val)
)
```
