```python
tmp_taman = pd.DataFrame()
tmp_museum = pd.DataFrame()
```

# Taman Surabaya

```python
tmp = taman_surabaya.copy()
tmp.drop(columns = ["id", "Keyword", "Website", "Plus_Code", "Reviews"], inplace=True)
tmp.dropna(subset = ["Name", "Address", "Rating"], inplace=True)
tmp["Provinsi"] = tmp["Address"].apply(lambda data: data.split(",")[::-1][0].strip())
tmp["Kota"] = tmp["Address"].apply(lambda data: data.split(",")[::-1][1].strip())
tmp["Kecamatan"] = tmp["Address"].apply(lambda data: data.split(",")[::-1][2].strip())
tmp.head()
```

```python
tmp["Provinsi"] = tmp["Provinsi"].apply(lambda data: " ".join(data.split()[:2]))
tmp["Provinsi"].replace("East Java", "Jawa Timur", inplace=True)
tmp["Provinsi"].astype("category")
```

```python
tmp.drop(tmp.loc[tmp["Kota"] == "Kabupaten Sidoarjo"].index, inplace=True)
tmp["Kota"].replace(["Kota SBY", "Surabaya City"], "Surabaya", inplace=True)
tmp["Kota"].astype("category")
```

```python
tmp["Kecamatan"] = tmp["Kecamatan"].apply(lambda data: " ".join(data.split()[1:]) if (len(data.split())>1 and "Kec" in data.split()[0]) else data)
tmp["Kecamatan"].astype("category")
```

```python
try:
    tmp["Rating"] = tmp["Rating"].str.replace(",", ".").astype(float)
except:
    print("Udah Benar")
tmp.head()
```

```python
print(tmp.isnull().sum())
tmp_new = pd.DataFrame()
tmp_new["nama"] = tmp["Name"]
tmp_new["kategori"] = "Taman"
tmp_new["rating"] = tmp["Rating"]
tmp_new["provinsi"] = tmp["Provinsi"]
tmp_new["kota"] = tmp["Kota"]
tmp_new["kecamatan"] = tmp["Kecamatan"]
tmp_new["alamat"] = tmp["Address"]
tmp_new["link"] = tmp["URL"]
tmp_taman = tmp_taman.append(tmp_new, ignore_index=True)
```

## Modelling dataset Taman

```python
# Mengimpor library TfidfVectorizer untuk pemodelan Content-Based Filtering
from sklearn.feature_extraction.text import TfidfVectorizer
 
# Inisialisasi TfidfVectorizer
tf = TfidfVectorizer()
 
# Melakukan perhitungan idf pada kolom kecamatan
tf.fit(taman['kecamatan'])
 
# Mendata penerbit buku pada data taman
tf.get_feature_names()

# Melakukan fit lalu ditransformasikan ke bentuk dataframe
tfidf_matrix = tf.fit_transform(taman['kecamatan'])
 
# Melihat ukuran dataframe tfidf
tfidf_matrix.shape

# Mengubah vektor tf-idf dalam bentuk matriks dengan fungsi todense()
tfidf_matrix.todense()

# Membuat dataframe untuk melihat tf-idf matrix
# Baris diisi dengan judul buku
# Kolom diisi dengan penerbit buku

pd.DataFrame(
    tfidf_matrix.todense(),
    columns=tf.get_feature_names(),
    index=taman['nama']
).sample(5, axis=0).sample(5, axis=1)
```

```python
# Mengimpor library cosine_similarity untuk pemodelan Content-Based Filtering
from sklearn.metrics.pairwise import cosine_similarity
 
# Menghitung cosine similarity pada matrix tf-idf
cosine_sim = cosine_similarity(tfidf_matrix)

# Menampilkan array hasil komputasi fungsi cosine_similarity
cosine_sim
```

```python
# Membuat dataframe dari variabel cosine_sim dengan baris dan kolom berupa nama
cosine_sim_df = pd.DataFrame(cosine_sim, index=taman['nama'], columns=taman['nama'])

# Menampilkan bentuk data cosine_sim_df
print('Bentuk :', cosine_sim_df.shape)
 
# Melihat similarity matrix pada setiap buku
cosine_sim_df.sample(5, axis=0).sample(5, axis=1)
```

```python
# Membuat fungsi rekomendasi data untuk Content-Based Filtering
def rekomendasi_taman(nama, similarity_data=cosine_sim_df, items=taman[['nama', 'kategori', 'rating', 'kecamatan', 'kota', 'provinsi', 'alamat', 'link']], k=20):
    """
    Rekomendasi data berdasarkan kemiripan dataframe
 
    Parameter:
    ---
    nama : tipe data string (str)
                Nama data (index kemiripan dataframe)
    similarity_data : tipe data pd.DataFrame (object)
                      Kesamaan dataframe, simetrik, dengan nama data sebagai 
                      indeks dan kolom
    items : tipe data pd.DataFrame (object)
            Mengandung kedua nama dan fitur lainnya yang digunakan untuk mendefinisikan kemiripan
    k : tipe data integer (int)
        Banyaknya jumlah rekomendasi yang diberikan
    ---
 
 
    Pada index ini, kita mengambil k dengan nilai similarity terbesar 
    pada index matrix yang diberikan (i).
    """

    # Melakukan matching data pertama yang paling mirip dengan parameter nama
    for i in taman["nama"]:
      count = 0
      for j in nama.lower().split():
        if j in i.lower().split():
          count += 1
      if count == len(nama.split()):
        nama = i
        break
    else:
      return "Tidak ada rekomendasi"

    # Mengambil data dengan menggunakan argpartition untuk melakukan partisi secara tidak langsung sepanjang sumbu yang diberikan    
    # Dataframe diubah menjadi numpy
    # Range(start, stop, step)
    index = similarity_data.loc[:,nama].to_numpy().argpartition(range(-1, -k, -1))
    
    # Mengambil data dengan similarity terbesar dari index yang ada
    closest = similarity_data.columns[index[-1:-(k+2):-1]]
    
    # Drop nama agar nama data yang dicari tidak muncul dalam daftar rekomendasi
    closest = closest.drop(nama, errors='ignore')
 
    return pd.DataFrame(closest).merge(items).head(k)
```

```python
# Membuat fungsi rekomendasi data untuk Content-Based Filtering
def rekomendasi_museum(nama, similarity_data=cosine_sim_df, items=museum[['nama', 'kategori', 'rating', 'kecamatan', 'kota', 'provinsi', 'alamat', 'link']], k=20):
    """
    Rekomendasi data berdasarkan kemiripan dataframe
 
    Parameter:
    ---
    nama : tipe data string (str)
                Nama data (index kemiripan dataframe)
    similarity_data : tipe data pd.DataFrame (object)
                      Kesamaan dataframe, simetrik, dengan nama data sebagai 
                      indeks dan kolom
    items : tipe data pd.DataFrame (object)
            Mengandung kedua nama dan fitur lainnya yang digunakan untuk mendefinisikan kemiripan
    k : tipe data integer (int)
        Banyaknya jumlah rekomendasi yang diberikan
    ---
 
 
    Pada index ini, kita mengambil k dengan nilai similarity terbesar 
    pada index matrix yang diberikan (i).
    """

    # Melakukan matching data pertama yang paling mirip dengan parameter nama
    for i in museum["nama"]:
      count = 0
      for j in nama.lower().split():
        if j in i.lower().split():
          count += 1
      if count == len(nama.split()):
        nama = i
        break
    else:
      return "Tidak ada rekomendasi"

    # Mengambil data dengan menggunakan argpartition untuk melakukan partisi secara tidak langsung sepanjang sumbu yang diberikan    
    # Dataframe diubah menjadi numpy
    # Range(start, stop, step)
    index = similarity_data.loc[:,nama].to_numpy().argpartition(range(-1, -k, -1))
    
    # Mengambil data dengan similarity terbesar dari index yang ada
    closest = similarity_data.columns[index[-1:-(k+2):-1]]
    
    # Drop nama agar nama data yang dicari tidak muncul dalam daftar rekomendasi
    closest = closest.drop(nama, errors='ignore')
 
    return pd.DataFrame(closest).merge(items).head(k)
```
